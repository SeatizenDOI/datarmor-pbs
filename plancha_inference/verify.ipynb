```python
from pathlib import Path
```

```python
for file in Path("./output/").iterdir():
    if "maurice_inference.o4190672" not in file.name: continue

    session_name, isFinish = "", False
    with open(file, "r") as f:
        for row in f:
            if "Launched session" in row:
                session_name = row.replace("\n", "").split(" ")[2]
            if "end succesfully !" in row:
                isFinish = True
    if not isFinish:
        print(file.name.split('.')[-1], session_name,"\t\t" ,isFinish)
```
