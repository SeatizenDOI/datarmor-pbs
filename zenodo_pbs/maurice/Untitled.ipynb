```python
from pathlib import Path
```

```python
cpt, cpt_wall, cpt_mission, cpt_convex = 0, 0, 0, 0
for file in Path(".").iterdir():
    isValid, isWallTime, isMissionStart, isConvex = False, False, False, False
    if "zen_up_julien.o4239490" not in file.name: continue
    session_name = ""
    with open(file, "r") as f:
        for row in f: 
            if "Working with session" in row:
                session_name = row.split(" ")[3].strip()
            if "End of process. On 1 sessions, 0 fails." in row:
                isValid = True
                cpt += 1
                break
            if "PBS: job killed: walltime" in row:
                isWallTime = True
                cpt_wall += 1
            if "AttributeError: 'float' object has no attribute 'split'" in row:
                isMissionStart = True
                cpt_mission += 1
            if "ValueError: Points cannot contain NaN" in row:
                isConvex = True
                cpt_convex += 1
                
    if not isValid:
        print(session_name)
    
    if not isValid and not isWallTime and not isMissionStart and not isConvex:
        print(f"Something strange with {file}")


        
print(97-cpt, cpt_wall, cpt_mission/2, cpt_convex)
```
