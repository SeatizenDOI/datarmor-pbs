```python
from pathlib import Path 
import shutil

RECIF_3D_FOLDER = Path("/media/bioeos/E/recif3D/2021-11-30")
RECIF_3D_FOLDER_session = Path("/media/bioeos/E/recif3D/sessions")

cpt = 0
for i, folder in enumerate(sorted(list(RECIF_3D_FOLDER.iterdir()))):
    
    if folder.is_file(): continue
    date = folder.name.split("T")[0].replace("acq_", "")
    cpt_str = str(cpt).rjust(2, '0')
    session_name = "20211130_REU-BOUCAN_AUV-1_01"
    cpt += 1
    if cpt not in [1,2,3,5,6,7,8,9]: continue
    print(session_name, cpt, folder)
    metadata = Path(RECIF_3D_FOLDER_session, session_name, "METADATA")
    dcim = Path(RECIF_3D_FOLDER_session, session_name, "DCIM")
    dcim.mkdir(exist_ok=True, parents=True)
    metadata.mkdir(exist_ok=True, parents=True)

    
    for file in folder.iterdir():
        
        if file.suffix.lower() != ".jpg":
            shutil.move(file, Path(metadata, file.name))
        else:
            shutil.move(file, Path(dcim, file.name))   
```

```python
from pathlib import Path 
import shutil

RECIF_3D_FOLDER = Path("/media/bioeos/E/recif3D/2021-12-02")
RECIF_3D_FOLDER_session = Path("/media/bioeos/E/recif3D/sessions")

cpt = 0
for i, folder in enumerate(sorted(list(RECIF_3D_FOLDER.iterdir()))):
    
    if folder.is_file(): continue
    date = folder.name.split("T")[0].replace("acq_", "")
    cpt_str = str(cpt).rjust(2, '0')
    session_name = "20211202_REU-BOUCAN_AUV-1_01"

    metadata = Path(RECIF_3D_FOLDER_session, session_name, "METADATA")
    dcim = Path(RECIF_3D_FOLDER_session, session_name, "DCIM")
    dcim.mkdir(exist_ok=True, parents=True)
    metadata.mkdir(exist_ok=True, parents=True)

    
    for file in folder.iterdir():
        
        if file.suffix.lower() != ".jpg":
            shutil.move(file, Path(metadata, file.name))
        else:
            shutil.move(file, Path(dcim, file.name))  
```

```python
import pandas as pd
HEADER = [
    "DiveNumber", "Date", "Time", "CameraModel", "SourceFormat", "ImageFilename", 
    "Latitude", "Longitude", "Depth", "Altitude", "Heading", "Roll_deg", "Pitch_deg", 
    "VelocityX", "VelocityY", "VelocityZ", "RotationRateX", "RotationRateY", "RotationRateZ", 
    "Pan", "Tilt", "Zoom", "Focus", "Iris", "Zoom_Reserved", "Focus_Reserved", 
    "Iris_Reserved", "EventName", "Tag", "Comment"
]
df_glob = pd.DataFrame()
for file in Path("/media/bioeos/E/recif3D/sessions/20211202_REU-BOUCAN_AUV-1_01/METADATA").iterdir():
    df = pd.read_csv(file, delimiter=";", names=HEADER)

    df_glob = pd.concat([df_glob, df])

list_images = df_glob["ImageFilename"].to_list()

for img in Path("/media/bioeos/E/recif3D/sessions/20211202_REU-BOUCAN_AUV-1_01/DCIM").iterdir():
    if img.name not in list_images:
        img.unlink()
```
