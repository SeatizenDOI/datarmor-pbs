```python
for session in sorted(list(SESSIONS.iterdir())):
    ortho_file = Path(session, "PROCESSED_DATA", "PHOTOGRAMMETRY_COLORED", "odm_orthophoto", "odm_orthophoto.tif")
    if not ortho_file.exists(): continue

    out_file = Path(OUTPUT_folder, f"{session.name}_ortho.tif")

    print(out_file)
    shutil.copy(ortho_file, out_file)
```
