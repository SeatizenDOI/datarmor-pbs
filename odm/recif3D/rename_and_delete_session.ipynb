```python
import shutil
from pathlib import Path
import pandas as pd
SESSIONS = Path("/scale/project/rummo/recif3D/sessions")
df = pd.read_csv("suivi_recif3D.csv")
```

```python
for i, row in df.iterrows():
    need_deleted = row["is_deleted"]
    old_session_path = Path(SESSIONS, row["old"])
    print(old_session_path, old_session_path.exists())
    if not old_session_path.exists(): continue
    if need_deleted:
        print(f"{old_session_path} will be deleted")
        shutil.rmtree(old_session_path)
        continue
    if not pd.isna(row['new']) :
        new_session_path = Path(SESSIONS, row['new'])
        print(f"{old_session_path} will be renamed to {new_session_path}")
        shutil.move(old_session_path, new_session_path)
```
